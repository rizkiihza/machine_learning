```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics.pairwise import pairwise_distances

from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.cluster.bang import bang, bang_visualizer

from mst_clustering import MSTClustering
```

### Import dataframe

```python
# iris 
iris = load_iris()
df_iris = pd.DataFrame(data=iris['data'], columns=iris['feature_names'])
df_iris.head()
```

```python
# tennis
df_tennis = pd.read_csv('tennis.csv')
df_tennis.drop(['play'], inplace=True, axis=1)

outlook = []
temp = []
humidity = []
windy = []
for i in range(len(df_tennis)):
    if(df_tennis['outlook'][i] == 'rainy'):
        outlook.append(0)
    elif(df_tennis['outlook'][i] == 'overcast'):
        outlook.append(1)
    elif(df_tennis['outlook'][i] == 'sunny'):
        outlook.append(2)
    if(df_tennis['temp'][i] == 'cool'):
        temp.append(0)
    elif(df_tennis['temp'][i] == 'mild'):
        temp.append(1)
    elif(df_tennis['temp'][i] == 'hot'):
        temp.append(2)
    if(df_tennis['humidity'][i] == 'normal'):
        humidity.append(0)
    elif(df_tennis['humidity'][i] == 'high'):
        humidity.append(1)
    if(not df_tennis['windy'][i]):
        windy.append(0)
    elif(df_tennis['windy'][i]):
        windy.append(1)
df_tennis['outlook'] = outlook
df_tennis['temp'] = temp
df_tennis['humidity'] = humidity
df_tennis['windy'] = windy
df_tennis.head()
```

### Transform Data To Numpy Array

```python
X_iris = np.array(df_iris)
X_tennis = np.array(df_tennis)
```

### KMeans

parameter yang digunakan:
    - iris:
        - n_cluster = 3
        - random_state = 0
    - tennis:
        - n_cluster = 2
        - random_state = 0

```python
# iris
kmeans_iris = KMeans(n_clusters = 3, random_state = 0)
Y_iris = kmeans_iris.fit_predict(X_iris)
print("Iris : ")
plt.figure(figsize=(8,5))
df_iris['cluster'] = Y_iris
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df_iris)
plt.show()
```

```python
# tennis
kmeans_tennis = KMeans(n_clusters = 2, random_state = 0)
Y_tennis = kmeans_tennis.fit_predict(X_tennis)
print("Tennis : ")
plt.figure(figsize=(8,5))
df_tennis["cluster"] = Y_tennis
sns.scatterplot(x='outlook', y='temp', hue='cluster', data=df_tennis)
plt.show()
```

### Agglomerative Clustering

parameter yang digunakan:
    - iris:
        - n_clusters = 3
        - memory = None
        - affinity = "euclidean"
        - linkage = "average"
    - tennis:
        - n_clusters = 2
        - memory = None
        - affinity = "euclidean"
        - linkage = "average"

```python
# iris
agglo_iris = AgglomerativeClustering(n_clusters = 3, memory = None, affinity="euclidean", linkage = "average")
Y_iris = agglo_iris.fit_predict(X_iris)
print("Iris : ")
plt.figure(figsize=(8,5))
df_iris['cluster'] = Y_iris
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df_iris)
plt.show()
```

```python
# tennis
agglo_tennis = AgglomerativeClustering(n_clusters = 2, memory = None, affinity = "manhattan", linkage='complete')
Y_tennis = agglo_tennis.fit_predict(X_tennis)
print("Tennis : ")
plt.figure(figsize=(8,5))
df_tennis["cluster"] = Y_tennis
sns.scatterplot(x='outlook', y='temp', hue='cluster', data=df_tennis)
plt.show()
```

### DBSCAN

parameter yang digunakan:
    - iris:
        - eps = 0.41
        - min_samples = 3
        - n_jobs = -1
        - metric = "euclidean"
    - tennis:
        - eps = 1
        - min_samples = 2
        - n_jobs = -1
        - metric = "manhattan"

```python
# iris
dbscan_iris = DBSCAN(eps = 0.41, min_samples=3, n_jobs = -1, metric="euclidean")
Y_iris = dbscan_iris.fit_predict(X_iris)
print("Iris : ")
plt.figure(figsize=(8,5))
df_iris['cluster'] = Y_iris
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df_iris)
plt.show()
```

```python
# tennis
dbscan_tennis = DBSCAN(eps = 1, min_samples=3, n_jobs=-1, metric="manhattan")
Y_tennis = dbscan_tennis.fit_predict(X_tennis)
print("Tennis : ")
plt.figure(figsize=(8,5))
df_tennis["cluster"] = Y_tennis
sns.scatterplot(x='outlook', y='temp', hue='cluster', data=df_tennis)
plt.show()
```

### Gaussian Mixture

parameter yang digunakan:
    - iris:
        - max_iter = 200
        - random_state = 0
        - n_components = 3
    - tennis:
        - max_iter = 200
        - random_state = 0
        - n_components = 2

```python
# iris
em_iris = GaussianMixture(max_iter=200, random_state = 0, n_components=3)
em_iris.fit(X_iris)
Y_iris = em_iris.predict(X_iris)
plt.figure(figsize=(8,5))
df_iris['cluster'] = Y_iris
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df_iris)
plt.show()
```

```python
# tennis
em_tennis = GaussianMixture(max_iter=200, random_state=0, n_components=2)
em_tennis.fit(X_tennis)
Y_tennis = em_tennis.predict(X_tennis)
print("Tennis : ")
plt.figure(figsize=(8,5))
df_tennis["cluster"] = Y_tennis
sns.scatterplot(x='outlook', y='temp', hue='cluster', data=df_tennis)
plt.show()
```

### KMedoids

paramter yang digunakan:
    - iris:
        - initial_index_medoids = [1, 90, 120]
        - tolerance = 0.02
        - ccore = False
    - tennis:
        - initial_index_medoids = [1, 13]
        - tolerance = 0.02
        - ccore = False

```python
# iris 
kmedoids_iris = kmedoids(X_iris, [1,90,120], tolerance = 0.02, ccore=False)
kmedoids_iris.process()
cluster = kmedoids_iris.get_clusters()

Y_iris = [0 for i in range(len(X_iris))]
for row in range(len(cluster)):
    for idx in cluster[row]:
        Y_iris[idx] = row
        
df_iris['cluster'] = Y_iris
plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df_iris)
plt.show()
```

```python
# Tennis
kmedoids_tennis = kmedoids(X_tennis, [1,13], tolerance = 0.02, ccore=False)
kmedoids_tennis.process()
cluster = kmedoids_tennis.get_clusters()

Y_tennis = [0 for i in range(len(X_tennis))]
for row in range(len(cluster)):
    for idx in cluster[row]:
        Y_tennis[idx] = row
        
df_tennis['cluster'] = Y_tennis
plt.figure(figsize=(8,5))
sns.scatterplot(x='outlook', y='temp', hue='cluster', data=df_tennis)
plt.show()
```

### MST Clustering

parameter yang digunakan:
     - iris:
         - cutoff = 3
     - tennis:
         - cutoff = 2

```python
mst_iris = MSTClustering(cutoff=3)
Y_iris = mst_iris.fit_predict(X_iris)
plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df_iris)
plt.show()
```

```python
mst_tennis = MSTClustering(cutoff=2)
Y_tennis = mst_tennis.fit_predict(X_tennis)
plt.figure(figsize=(8,5))
sns.scatterplot(x='outlook', y='temp', hue='cluster', data=df_tennis)
plt.show()
```

### BANG (Grid Clustering)

parameter yang digunakan:
    - iris:
        - levels = 3
    - tennis:
        - levels = 2

```python
bang_iris = bang(X_iris.tolist(), levels=14)
bang_iris.process()

clusters = bang_iris.get_clusters()

Y = [0 for i in range(len(X_iris))]
for row in range(len(clusters)):
    for index in clusters[row]:
        Y[index] = row
        
df_iris['cluster'] = Y
plt.figure(figsize=(8,5))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='cluster', data=df_iris)
plt.show()
```

```python
bang_tennis = bang(X_tennis.tolist(), 7)
bang_tennis.process()

clusters = bang_tennis.get_clusters()

Y = [0 for i in range(len(X_tennis))]
for row in range(len(clusters)):
    for index in clusters[row]:
        Y[index] = row
        
df_tennis['cluster'] = Y
plt.figure(figsize=(8, 5))
sns.scatterplot(x='outlook', y='temp', hue='cluster', data=df_tennis)
plt.show()
```

### Save model to pickle

```python
import pickle

# Kmeans
filename = "models/iris/kmeans_iris.sav"
pickle.dump(kmeans_iris, open(filename, 'wb'))

filename = "models/tennis/kmeans_tennis.sav"
pickle.dump(kmeans_tennis, open(filename, 'wb'))

# Agglo
filename = "models/iris/agglo_iris.sav"
pickle.dump(agglo_iris, open(filename, 'wb'))

filename = "models/tennis/agglo_tennis.sav"
pickle.dump(agglo_tennis, open(filename, 'wb'))

# DBSCAN
filename = "models/iris/dbscan_iris.sav"
pickle.dump(dbscan_iris, open(filename, 'wb'))

filename = "models/tennis/dbscan_tennis.sav"
pickle.dump(dbscan_tennis, open(filename, 'wb'))

# EM GaussianMixture
filename = "models/iris/em_iris.sav"
pickle.dump(em_iris, open(filename, 'wb'))

filename = "models/tennis/em_tennis.sav"
pickle.dump(em_tennis, open(filename, 'wb'))

# KMedoids
filename = "models/iris/kmedoids_iris.sav"
pickle.dump(kmedoids_iris.get_cluster_encoding(), open(filename,'wb'))

filename = "models/tennis/kmedoids_tennis.sav"
pickle.dump(kmedoids_tennis.get_cluster_encoding(), open(filename,'wb'))

# MST
filename = "models/iris/mst_iris.sav"
pickle.dump(mst_iris, open(filename, 'wb'))

filename = "models/tennis/mst_tennis.sav"
pickle.dump(mst_tennis, open(filename, 'wb'))

# BANG (Grid Clustering)
filename = "models/iris/bang_iris.sav"
pickle.dump(bang_iris, open(filename, 'wb'))

filename = "models/tennis/bang_tennis.sav"
pickle.dump(bang_tennis, open(filename, 'wb'))
```
